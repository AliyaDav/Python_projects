```python
import pandas as pd

df = pd.read_excel('US_stocks.xlsx', usecols=lambda x: 'Unnamed' not in x,)
print (df)
```

```python
tickers = df.Symbol.to_list()
tickers[0:5]
```

```python
data = pd.DataFrame(columns=tickers)

# Fetch the data

for ticker in tickers:
    data[ticker] = yf.download(tickers,'2020-01-24','2020-04-24')['Adj Close']

# Print first 5 rows of the data
data.head()
```

```python
data = yf.download(tickers[0:5],'2020-01-24','2020-04-24')['Adj Close']
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt

# Plot all the close prices
#data.plot(figsize=(10, 7))
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Adjusted Close Price.Basic materials", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)


# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
